```python
from pathlib import Path
import pandas as pd
```

```python
def create_ogs(main_folder):
    rank_dfs = {}
    
    for rank_folder in main_folder.iterdir():
        if rank_folder.is_dir():
            specific = rank_folder.name.split('-')
            name = specific[0]
            
            temp = []
            for csv_file in rank_folder.glob("*.csv"):
                df = player_processing(csv_file)
                temp.append(df)
            merged_df = pd.concat(temp, ignore_index=True)
            merged_df['rank'] = name
            rank_dfs[name] = merged_df
    return rank_dfs


def export_ogs(rank_dfs):
    output_folder = Path(f"processed_csv's")
    output_folder.mkdir(exist_ok=True)
    for rank, df in rank_dfs.items():
        output_file = output_folder / f"{rank}.csv"
        df.to_csv(output_file, index=False)
```

```python
def player_processing(rank_file):
    df = pd.read_csv(rank_file)
    skill = player_processing_helper(df)
    temp = df.columns[0]
    col_names = temp.split(';')
    col_names.pop(1)
    return pd.DataFrame(data=skill, columns=col_names)

def player_processing_helper(df):
    matrix = []
    for i in df.index:
        first = df.iloc[i, 0]
        data = first.split(';;')
        color = data[0]
        rest = data[1].split(';')
        data.pop()
        data = data + rest
        matrix.append(data)
    return matrix
```

```python
def main():
    main_folder = Path(r"replay_csvs")

    rank_dfs = create_ogs(main_folder)
    export_ogs(rank_dfs)
```
